```python
#@title multi-processing -FF
#@markdown FF는 실행 시간이 짧아서 multi-processing이 속도가 느릴 수 있음

# multi-processing

'''
rdkit get mol block
'''

@ray.remote
def parallel_smiles2mol(row, bar):
  ray.get(bar.update.remote(1))
  try:
    return smiles2mol(*row[["SMILES"]].values, N_conformers=100)
  except:
    return None

##### FORCE FIELD #######

'''
rdkit
'''

@ray.remote
def parallel_rdkit_MMFF94_optimize(row, bar):
  ray.get(bar.update.remote(1))
  return rdkit_optimize(*row[["mol", "charge"]].values, forcefield="MMFF", mmffVariant="MMFF94", smiles_input=False)

@ray.remote
def parallel_rdkit_MMFF94s_optimize(row, bar):
  ray.get(bar.update.remote(1))
  return rdkit_optimize(*row[["mol", "charge"]].values, forcefield="MMFF", mmffVariant="MMFF94s", smiles_input=False)

@ray.remote
def parallel_rdkit_UFF_optimize(row, bar):
  ray.get(bar.update.remote(1))
  return rdkit_optimize(*row[["mol", "charge"]].values, forcefield="UFF", smiles_input=False)


'''
openbabel
'''

# openbabel이 binding인지 wrapping인지 모르겠음 안되면 serial로 바꿔야 함

@ray.remote
def parallel_openbabel_UFF_optimize(row, bar):
  ray.get(bar.update.remote(1))
  return openbabel_optimize(*row[["mol", "charge"]].values, forcefield="uff", smiles_input=False)

@ray.remote
def parallel_openbabel_MMFF94_optimize(row, bar):
  ray.get(bar.update.remote(1))
  return openbabel_optimize(*row[["mol", "charge"]].values, forcefield="mmff94", smiles_input=False)

@ray.remote
def parallel_openbabel_MMFF94s_optimize(row, bar):
  ray.get(bar.update.remote(1))
  return openbabel_optimize(*row[["mol", "charge"]].values, forcefield="mmff94s", smiles_input=False)

@ray.remote
def parallel_openbabel_GHEMICAL_optimize(row, bar):
  ray.get(bar.update.remote(1))
  return openbabel_optimize(*row[["mol", "charge"]].values, forcefield="ghemical", smiles_input=False)

@ray.remote
def parallel_openbabel_GAFF_optimize(row, bar):
  ray.get(bar.update.remote(1))
  return openbabel_optimize(*row[["mol", "charge"]].values, forcefield="gaff", smiles_input=False)
```

```python
#@title run job function

def run_parallel_job(DATA_TABLE, calculator, col_name):
  '''

  '''
  # ray setting
  ray.init(ignore_reinit_error=True)
  remote_tqdm = ray.remote(tqdm_ray.tqdm)
  NUM_OF_DATA_TABLE_ROW = DATA_TABLE.shape[0]
  bar = remote_tqdm.remote(total=NUM_OF_DATA_TABLE_ROW)

  # 병렬 작업 ID 획득
  task_id = []
  for _, row in DATA_TABLE.iterrows():
    task_id.append(calculator.remote(row, bar))

  # 실제 결과 획득
  __result__ = ray.get(task_id)

  # finalize ray
  ray.get(bar.close.remote())  # Close the progress bar properly
  ray.shutdown() # shutdown Ray

  # save result
  DATA_TABLE[col_name] = __result__
  DATA_TABLE[["CSID", col_name]].to_csv(f"{col_name}.csv", index=False)

  clear_output()

def run_serial_job(DATA_TABLE, calculator, col_name):
  '''

  '''
  # 결과 획득
  __result__ = []
  N_total = DATA_TABLE.shape[0]
  for _, row in tqdm(DATA_TABLE.iterrows(), total = N_total):
    __result__.append(calculator(row))

  # save result
  DATA_TABLE[col_name] = __result__
  DATA_TABLE[["CSID", col_name]].to_csv(f"{col_name}.csv", index=False)

  clear_output()
```
